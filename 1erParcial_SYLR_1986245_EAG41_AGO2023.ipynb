```python
#Importar las librerias
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

##1. Películas

```python
#1. Cargar base de datos
peliculas = pd.read_csv("https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/peliculas.csv")
peliculas.head(5)
```

```python
#a) Total de peliculas
total_pel = peliculas['title'].count()
print(f'Hay un total de', total_pel, 'peliculas')
```

```python
#a) Total de géneros
total_gen = peliculas['category'].value_counts()
cant_gen = len(total_gen)
print(f'Hay un total de', cant_gen, 'géneros')
```

```python
#a) Promedio de duración de películas
prom = peliculas['length'].mean()
print(f'El promedio de la duración de las películas es de', prom)
```

```python
#b) Crear el dataframe de documentales con duración mayor a 2 horas
peliculas_dormir = peliculas.query('category=="Documentary" & length > 120')
peliculas_dormir.head(5)
```

```python
#c) Peliculas de cada género
total_gen_pel = peliculas['category'].value_counts()
total_gen_pel
```

```python
#d) Promedio de las peliculas por género
prom_gen_pel = peliculas.groupby('category')['length'].mean()
prom_gen_pel.round(1)
```

```python
#e) Duracion corta, normal o larga
condicion_dur = [peliculas['length'] < 90,
                 peliculas['length'] < 150,
                 peliculas['length'] > 150]
valor_cond = ['Corta', 'Normal', 'Larga']
peliculas['Condicion de duración'] = np.select(condicion_dur,valor_cond, default='Invalido')
peliculas.head(10)
```

##2. Población

```python
#2. Cargar base de datos
poblacion_Asgard = pd.read_csv("https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/poblacion_Asgard.csv")
poblacion_Asgard.head(5)
```

```python
#a) Ordenando alfabeticamente los estados y municipios
ordenar = poblacion_Asgard.sort_values(by=['Estado'], ascending=True)
ordenar
```

```python
#a) Ordenando alfabeticamente los municipios
ordenar = poblacion_Asgard.sort_values(by=['Municipio'], ascending=True)
ordenar
```

```python
#b) Total de estados
total_estados = poblacion_Asgard['Estado'].nunique()
print(f'Total de estados:', total_estados)
```

```python
#b) Municipios en cada estado
municipio_estado = poblacion_Asgard.groupby('Estado')['Municipio'].nunique()
print(f'Total de municipios por Estado:')
municipio_estado
```

```python
#b) Total de municipios
total_municipios = poblacion_Asgard['Municipio'].nunique()
print(f'Total de municipios:', total_municipios)
```

```python
#c) Población total del país
poblacion_total = poblacion_Asgard['Población'].sum()
print(f'La poblacion total del país de Asgard es de', poblacion_total, 'habitantes')
```

```python
#c) Población total de cada estado
poblacion_total_estado = poblacion_Asgard.groupby('Estado')['Población'].sum()
print(f'Total de población por Estado:')
poblacion_total_estado
```

```python
#d) Mediana de la población de los municipios
mediana_municipios = poblacion_Asgard['Población'].median()
ent_mediana_municipios = int(mediana_municipios)
ent_mediana_municipios
```

```python
#d) Clasificarlos muy poblados o poco poblados
poblacion_Asgard['Mediana de población']=np.where(poblacion_Asgard['Población'] > ent_mediana_municipios, 'Muy poblados','Poco poblados')
poblacion_Asgard.head(10)
```

```python
#e) Cinco municipios más pobaldos
pob_Asg1 = pd.DataFrame(poblacion_Asgard)
#Ordenar los datos por Población de forma descendente
ordenar_df = pob_Asg1.sort_values(by='Población', ascending=False)
#Obtener los cinco municipios más poblados
primeros_cinco = ordenar_df.head(5)
print(primeros_cinco)
```

```python
#e) Cinco municipios menos poblados
pob_Asg2 = pd.DataFrame(poblacion_Asgard)
#Ordenar los datos por Población de forma ascendente
ordenar_df = pob_Asg2.sort_values(by='Población', ascending=True)
#Obtener los cinco municipios menos poblados
ultimos_cinco = ordenar_df.head(5)
print(ultimos_cinco)
```

```python
#f) Municipio más poblado de cada estado
pob_Asg3= pd.DataFrame(poblacion_Asgard)
#Encontar el Municipio más poblado de cada Estado
municipios_mas_poblados = pob_Asg3.groupby('Estado')['Población'].idxmax()
mas_poblados = pob_Asg3.loc[municipios_mas_poblados]
print(mas_poblados)
```

## 3. Grafico de barras

```python
#a) Los 15 municipios más poblados de Asgard
pob_Asg4 = pd.DataFrame(poblacion_Asgard)
# Ordenar los municipios por población
pob_Asg4_ordenado = pob_Asg4.sort_values(by='Población', ascending=False)
# Obtener los 15 municipios más poblados
quince_municipios = pob_Asg4_ordenado.head(15)
# Crear la gráfica de barras
plt.bar(quince_municipios['Municipio'], quince_municipios['Población'])
plt.xlabel('Municipio')
plt.ylabel('Población')
plt.title('15 Municipios Más Poblados de Asgard')
plt.xticks(rotation=90)
plt.show()
```
