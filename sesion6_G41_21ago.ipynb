```python
#Importar las librerias
import numpy as np
import pandas as pd
import matplotlib .pyplot as plt
%matplotlib inline
```

```python
nacimientos = pd.DataFrame({'Municipios':['Monterrey','Apodaca','Escobedo','Santiago'], 'Bebes_Nacidos':[233, 569, 780, 123], 'Fallecidos': [123, 80, 23, 12]})
nacimientos
```

```python
#Ordenar la grafica de barras
nacimientos = nacimientos.sort_values('Bebes_Nacidos',ascending=False)
```

```python
plt.bar(nacimientos['Municipios'], nacimientos['Bebes_Nacidos'])
plt.title('Nacimiento en Nuevo León en 2020')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de bebes')
plt.show()
```

```python
personas =[ 'Juan', 'Pedro', 'Ana', 'Ale']
edades =[ 23, 18, 30, 27]
```

```python
plt.bar(personas,edades)
plt.show()
```

```python
#Fig es nombre del lienzo
fig, (panel1, panel2) = plt.subplots(2,1)

#Grafica1 "NACIMIENTOS"
x1 = nacimientos['Municipios']
y1 = nacimientos['Bebes_Nacidos']
panel1.bar(x1, y1)

#Grafica2 "EDADES"
x2 = personas
y2 = edades
panel2.bar(x2, y2)
plt.show()
```
